### Analysis on Daily Data set

#### Importing dataset and getting insights
                                           

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
     
```

```python
df=pd.read_csv("daily_data.csv")
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python


# making 'DATE' column is in datetime format
df['DATE'] = pd.to_datetime(df['DATE'])

# Plot using sns.lineplot
sns.lineplot(x="DATE", y="DailyPrecipitation", data=df)
plt.title("Daily Precipitation Events")
plt.xlabel("Date")
plt.ylabel("Precipitation (mm)")
plt.xticks(rotation=45)
plt.show()

     
```

```python
df['DailyMaximumDryBulbTemperature'] = pd.to_numeric(df['DailyMaximumDryBulbTemperature'], errors='coerce')
heatmap_data = df.dropna().pivot(index="DATE", columns="STATION", values="DailyMaximumDryBulbTemperature")
sns.heatmap(heatmap_data, cmap="YlOrRd", cbar_kws={'label': 'Temperature (°C)'})
plt.title("Heatmap of Daily Maximum Temperature")
plt.xlabel("Station")
plt.ylabel("Date")
plt.xticks(rotation=45)
print(heatmap_data.shape)
plt.show()
     
```

### Analysis on Monthly Dataset

```python
df1=pd.read_csv("monthly_data.csv")
df1
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1.isnull().sum()
```

```python


# Ensure 'DATE' columns are in datetime format
df['DATE'] = pd.to_datetime(df['DATE'])
df1['DATE'] = pd.to_datetime(df1['DATE'])


# Plot the temperature trends
sns.lineplot(x="DATE", y="MonthlyMeanTemperature", data=df1, label="Monthly Mean Temperature")
sns.lineplot(x="DATE", y="DailyAverageDryBulbTemperature", data=df, label="Daily Average Dry Bulb Temperature")
plt.title("Temperature Trends over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.show()
```

```python
numerical_columns = ['MonthlyMaximumTemperature', 'MonthlyMinimumTemperature', 'MonthlyMeanTemperature','MonthlyTotalLiquidPrecipitation', 'MonthlyGreatestPrecip', 'MonthlyGreatestSnowDepth']
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df1[column], kde=True)
    plt.title(column)
    plt.xlabel(column)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x="DATE", y="MonthlyMeanTemperature", data=df1, color = "BROWN")
plt.title("Monthly Mean Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature in Celsius")
plt.xticks(rotation=45)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
temperature_series = df1['MonthlyMeanTemperature']
if len(temperature_series) >= 24:
    result = seasonal_decompose(temperature_series, model='additive', period=12)
    result.plot()
    plt.show()
else:
    print("Insufficient data for seasonal decomposition. Need at least 24 observations.")
     
```

### Analysis on Three Hour dataset


```python
df3=pd.read_csv("three_hour_data.csv")
df3.head()
```

```python
df3.tail()
```

```python
df3.isnull().sum()
```

```python
df3.info()
```

```python
df3.describe()
```

```python

numerical_columns = ['HourlyDewPointTemperature', 'HourlyDryBulbTemperature', 'HourlyRelativeHumidity', 'HourlyWindSpeed']
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df3[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df3[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
sns.lineplot(x='DATE', y='HourlyDryBulbTemperature', data=df3, label='Hourly Dry Bulb Temperature')
sns.lineplot(x='DATE', y='HourlyWindSpeed', data=df3, label='Hourly Wind Speed')
plt.title('Time Trends of Hourly Dry Bulb Temperature and Wind Speed')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
